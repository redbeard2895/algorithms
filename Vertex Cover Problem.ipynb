This Jupyter notebook implments a <b>Polynomial Time Approximate algorithm</b> for the <b>Vertex Cover problem</b> which is given as:

A vertex cover of an undirected graph G = (V, E) is a subset V'⊆V such that if (u,v)∈ E, then either u∈V' or v∈V' or both. That is, each vertex 'covers' its incident edges, and a vertex cover for G is a set of vertices that covers all the edges in E.

##### This problem is known to be NP Complete as of today.

This Algorithm does not give the Optimal solution but rather an Approximate solution.

### Importing libraries

```python
import networkx as nx              # for plotting the graph
import matplotlib.pyplot as plt    # for displaying the plotted graph
```

### User defined class - graph

   This class is used for maintaining a graph. It takes the dictionary of Vertices and Edges during object creation.
   It has the following functions:
    1. del_edges() : Accepts an edge as input and deletes it from the graph
    2. visualize(): Displays the current state of the graph

```python
class graph:
    
    def __init__(self, V_E):
        self.V = list(V_E.keys())
        self.E = list()
        for key, val in V_E.items():
            for edge in val:
                self.E.append({key, edge})
                
    def del_edges(self, edge):
        edges = self.E
        u, v = edge
        for e in edges:
            if u in e:
                edges.remove(e)
            elif v in e:
                edges.remove(e)
        
    def visualize(self):
        fig = nx.Graph()
        fig.add_edges_from(self.E)
        nx.draw_networkx(fig)
        plt.show()                      
```

```python
v_e = {'a' : ['b'],
       'b' : ['c'],
       'c' : ['b', 'd', 'e'],
       'd' : ['c', 'e', 'f'],
       'e' : ['c', 'd', 'f'],
       'f' : ['d', 'e'],
       'g' : ['e']}
```

```python
G = graph(v_e)    # creating the graph
```

```python
G.visualize()
```

```python
C = set()
E_ = G.E
while len(E_) != 0:
    temp = E_.pop()
    C = C.union(temp)
    G.del_edges(temp)
    G.visualize()  
```
